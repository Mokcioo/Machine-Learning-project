```python
import json
import matplotlib.pyplot as plt
```

#### Red - validation || blue - training

```python
dict = json.load(open('training_history.json', 'r'))
yAxis = dict['loss']
yAxis2 = dict['val_loss']
epochs = [i for i in range(1,13)]
plt.plot(epochs,yAxis, color='blue')
plt.plot(epochs,yAxis2, color='red')
plt.xlabel('epochs')
plt.ylabel('loss')
plt.show()
```

```python
dict = json.load(open('training_history.json', 'r'))
yAxis = dict['accuracy']
yAxis2 = dict['val_accuracy']
epochs = [i for i in range(1,13)]
plt.plot(epochs,yAxis, color='blue')
plt.plot(epochs,yAxis2, color='red')
plt.xlabel('epochs')
plt.ylabel('accuracy')
plt.show()
```
